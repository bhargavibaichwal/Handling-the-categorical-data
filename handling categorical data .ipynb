## dictionary comprehension

```python
labels=flights['carrier'].astype('category').cat.categories.tolist()
labels
replace_map_comp={'carrier':{k:v for k ,v in zip(labels,list(range(1,len(labels)+1)))}}
flights_replace1=c.copy()
flights_replace1.replace(replace_map_comp, inplace=True)
flights_replace1
#encoding the data 
```

## type casting (changing carrier and origin to category )

```python
flights_lc=c.copy()
flights_lc['carrier']=flights_replace1['carrier'].astype('category')
flights_lc['origin']=flights_replace1['origin'].astype('category')
print(flights_lc.dtypes)
```

```python
#counting the number of flights in carrier foorm origin
c.groupby(['origin','carrier']).count()
```

## prove that data frame with category dtype is faster than object type
#calculate the same code on category calculate the time and then calculate the same on objects

```python
import time
%timeit flights_lc.groupby(['origin','carrier']).count()#category changed file
```

```python
import time
%timeit c.groupby(['origin','carrier']).count()# object original file
```

#### hence category type is faster than object type
